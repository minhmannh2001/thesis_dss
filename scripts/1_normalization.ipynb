```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
try:
    sqliteConnection = sqlite3.connect('./../data/sqlite/thesis-DSS.db')
    print("Database created and Successfully Connected to SQLite")

    # Create thesis dataframe from table thesis
    thesis_df = pd.read_sql_query('SELECT * FROM thesis;', sqliteConnection)
    print('Create thesis dataframe successfully')

    # Create teacher dataframe from table teacher
    teacher_df = pd.read_sql_query('SELECT * FROM teacher;', sqliteConnection)
    print('Create teacher dataframe successfully')

except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if sqliteConnection:
        sqliteConnection.close()
        print("The SQLite connection is closed")
```

```python
no_occurrence_of_professor = np.array(no_occurrence_of_professor)
no_occurrence_of_doctor = np.array(no_occurrence_of_doctor)
no_occurrence_of_masters = np.array(no_occurrence_of_masters)
no_occurrence_of_engineering_degree = np.array(no_occurrence_of_engineering_degree)
no_occurrence_of_bachelor = np.array(no_occurrence_of_bachelor)

degrees_point = no_occurrence_of_doctor * 5 + no_occurrence_of_doctor * 4 + no_occurrence_of_masters * 3 + no_occurrence_of_engineering_degree * 2 + no_occurrence_of_bachelor * 1
```

```python
min_degrees_point = teacher_df['degrees_point'].min()
max_degrees_point = teacher_df['degrees_point'].max()
```

```python
normalized_degrees_point_attribute = []

for index, row in teacher_df.iterrows():
    normalized_degrees_point_attribute.append((row['degrees_point'] - min_degrees_point) / (max_degrees_point - min_degrees_point))

teacher_df['normalized_degrees_point_attribute'] = normalized_degrees_point_attribute
```

### Normalize fields field

```python
teacher_df['major'] = teacher_df['major'].replace('Khoa Kỹ thuật Máy tính', 'Kỹ thuật Máy tính')
teacher_df['major'] = teacher_df['major'].replace('Kỹ thuật máy tính', 'Kỹ thuật Máy tính')
teacher_df['major'] = teacher_df['major'].replace('Khoa Kỹ thuật máy tính', 'Kỹ thuật Máy tính')
```

```python
unique_majors = teacher_df.major.unique()
unique_majors
```

```python
import operator
from collections import defaultdict
from scipy.spatial import distance
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
all_sentences = []
for index, row in teacher_df.iterrows():
    splitted_terms = row['fields'].split(',')
    lowered_terms = [term.lower() for term in splitted_terms]
    clean_terms = [term.strip() for term in lowered_terms]
    all_sentences.extend(clean_terms)
```

```python
len(all_sentences)
```

```python
vocab = []

for sentence in all_sentences:
    sentence = sentence.split()
    for word in sentence:
        if word not in vocab:
            vocab.append(word)

# number of words in the vocab
len_vector = len(vocab)
```

```python
len_vector
```

```python
#Index dictionary to assign an index to each word in vocabulary
index_word = {}
i = 0
for word in vocab:
    index_word[word] = i 
    i += 1
```

```python
def bag_of_words(sent):
    count_dict = defaultdict(int)
    vec = np.zeros(len_vector)
    for item in sent:
        count_dict[item] += 1
    for key, item in count_dict.items():
        if key in index_word:
            vec[index_word[key]] = item
    return vec
```

```python
career_paths = {}

for major in unique_majors:
    career_paths[major] = {}

for index, row in teacher_df.iterrows():
    splitted_terms = row['fields'].split(',')
    splitted_terms = [term.lower() for term in splitted_terms]
    splitted_terms = [term.strip() for term in splitted_terms]

    for term in splitted_terms:
        if len(career_paths[row['major']]) == 0:
            career_paths[row['major']][term] = 1
        else:
            keys = list(career_paths[row['major']].keys())
            for key in keys:
                similarity_score = 1 - distance.cosine(bag_of_words(term.split()), bag_of_words(key.split()))
                if similarity_score > 0.8:
                    career_paths[row['major']][key] += 1
                else:
                    career_paths[row['major']][term] = 1
```

```python
for major in unique_majors:
    print('Major: ', major)
    print(max(career_paths[major].items(), key=operator.itemgetter(1))[1])
    for career_path in career_paths[major]:
        print(career_path)
    print()
```

```python
max(career_paths['Hệ thống Thông tin'].items(), key=operator.itemgetter(1))[1]
```

```python
related_majors = []

for index, row in thesis_df.iterrows():
    major_scores = {}
    for major in unique_majors:
        major_scores[major] = 0
        for term in career_paths[major]:
            similarity_score_title = 1 - distance.cosine(bag_of_words(term.split()), bag_of_words(row['title'].split()))
            similarity_score_description = 1 - distance.cosine(bag_of_words(term.split()), bag_of_words(row['description'].split()))
            major_scores[major] = major_scores[major] + similarity_score_title + similarity_score_description
    related_majors.append(max(major_scores.items(), key=operator.itemgetter(1))[0])
```

```python
thesis_df['related_major'] = related_majors
```

### Normalize teacher's age

```python
import re
```

```python
min_graduation_years = []

for index, row in teacher_df.iterrows():
    graduation_years = re.findall('\d{4}', row['degrees'])
    graduation_years = [int(year) for year in graduation_years]
    min_graduation_year = min(graduation_years)
    min_graduation_years.append(min_graduation_year)

teacher_df['min_graduation_year'] = min_graduation_years
```

```python
teacher_df
```

```python
min_graduation_year =  teacher_df['min_graduation_year'].min()
max_graduation_year =  teacher_df['min_graduation_year'].max()
```

```python
normalized_age_attribute = []

for index, row in teacher_df.iterrows():
    normalized_age_attribute.append((row['min_graduation_year'] - min_graduation_year) / (max_graduation_year - min_graduation_year))

teacher_df['normalized_age_attribute'] = normalized_age_attribute
```

### Normalize no_researches and no_awards

```python
min_no_researches =  teacher_df['no_researches'].min()
max_no_researches =  teacher_df['no_researches'].max()
```

```python
normalized_no_researches_attribute = []

for index, row in teacher_df.iterrows():
    normalized_no_researches_attribute.append((row['no_researches'] - min_no_researches) / (max_no_researches - min_no_researches))

teacher_df['normalized_no_researches_attribute'] = normalized_no_researches_attribute
```

```python
min_no_awards =  teacher_df['no_awards'].min()
max_no_awards =  teacher_df['no_awards'].max()
```

```python
normalized_no_awards_attribute = []

for index, row in teacher_df.iterrows():
    normalized_no_awards_attribute.append((row['no_awards'] - min_no_awards) / (max_no_awards - min_no_awards))

teacher_df['normalized_no_awards_attribute'] = normalized_no_awards_attribute
```
