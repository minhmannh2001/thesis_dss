```python
import pandas as pd
```

```python
thesis_files = [
    './../data/Cơ sở dữ liệu - Đồ án - Lâm.csv',
    './../data/Cơ sở dữ liệu - Đồ án - Mạnh.csv',
    './../data/Cơ sở dữ liệu - Đồ án - Thắng.csv'
]
```

```python
thesis_dfs = [pd.read_csv(thesis_file) for thesis_file in thesis_files]
thesis_df = pd.concat(thesis_dfs)
thesis_df.reset_index(drop=True, inplace=True)
```

```python
thesis_df.info()
```

```python
thesis_df[['Description', 'Degree', 'Co-researchers']] = thesis_df[['Description', 'Degree', 'Co-researchers']].fillna(value='')
thesis_df['ThesisID'] = range(1, len(thesis_df) + 1)
thesis_df
```

```python
thesis_df.info()
```

```python
import re

thesis_degrees = []

for index, row in thesis_df.iterrows():
    thesis_degree = row['Degree']
    if thesis_degree != '':
        thesis_degree = thesis_degree.split(',')
        thesis_degree = [item.strip() for item in thesis_degree]
        thesis_degrees.extend(thesis_degree)

thesis_degrees = list(set(thesis_degrees))
thesis_degrees.remove(': ThSKH')
thesis_degrees.append('ThSKH')
thesis_degrees
```

```python
teacher_files = [
    './../data/Cơ sở dữ liệu - Giáo viên - Lâm.csv',
    './../data/Cơ sở dữ liệu - Giáo viên - Mạnh.csv',
    './../data/Cơ sở dữ liệu - Giáo viên - Thắng.csv'
]
```

```python
teacher_dfs = [pd.read_csv(teacher_file) for teacher_file in teacher_files]
teacher_df = pd.concat(teacher_dfs)
teacher_df.reset_index(drop=True, inplace=True)
```

```python
teacher_df.info()
```

```python
teacher_df[['NumberOfResearches', 'NumberOfAwards']] = teacher_df[['NumberOfResearches', 'NumberOfAwards']].fillna(value=0)
teacher_df
```

```python
teacher_df = teacher_df.astype({'NumberOfAwards': int, 'NumberOfResearches': int})
```

```python
teacher_df.info()
```

```python
teacher_df = teacher_df.drop('Subjects', axis=1)
teacher_df
```

```python
teacher_df['TeacherID'] = range(1, len(teacher_df) + 1)
teacher_df
```

```python
teacher_ids = []

for index, row in thesis_df.iterrows():
    teacher_ids.append(teacher_df[teacher_df['TeacherName'] == row['TeacherName']].TeacherID.values[0])

thesis_df = thesis_df.rename(columns={'TeacherName': 'TeacherID'})
thesis_df['TeacherID'] = teacher_ids
```

```python
thesis_df
```

```python
import sqlite3
from sqlite3 import Error


def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn
```

```python
database = './../data/sqlite/thesis-DSS.db'
conn = create_connection(database)
```

```python
def create_table(conn, create_table_sql):
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
sql_create_projects_table = """CREATE TABLE IF NOT EXISTS thesis (
                                    pk_thesis_id INTEGER PRIMARY KEY,
                                    fk_teacher_id INTEGER NOT NULL,
                                    title TEXT NOT NULL,
                                    description TEXT,
                                    degree TEXT,
                                    coresearchers TEXT,
                                    FOREIGN KEY (fk_teacher_id) REFERENCES teacher (pk_teacher_id)
                                );
                                """

sql_create_tasks_table = """CREATE TABLE IF NOT EXISTS teacher (
                                pk_teacher_id INTEGER PRIMARY KEY,
                                teacher_name text NOT NULL,
                                major TEXT NOT NULL,
                                degrees TEXT NOT NULL,
                                fields TEXT NOT NULL,
                                interested_fields TEXT NOT NULL,
                                no_researches INTEGER NOT NULL,
                                no_awards INTEGER NOT NULL
                            );"""

# create a database connection
conn = create_connection(database)

# create tables
if conn is not None:
    # create projects table
    create_table(conn, sql_create_projects_table)

    # create tasks table
    create_table(conn, sql_create_tasks_table)
else:
    print("Error! cannot create the database connection.")
```
